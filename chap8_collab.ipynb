```python
class DotProduct(Module):
    def __init__(self,user,movie,factors,y_range):
        self.user_factor = Embedding(user,factors)
        self.movie_factor = Embedding(movie,factors)
        self.y_range = y_range
    
    def forward(self,x):
        user = self.user_factor(x[:,0])
        movie = self.movie_factor(x[:,1])
        res =  (user * movie).sum(dim=1)
        return sigmoid_range(res,*self.y_range)
```

## Creating a model with user and movie biases

```python
class DotProductBias(Module):
    def __init__(self,user,movie,factors,y_range):
        self.user_factor = Embedding(user,factors)
        self.user_bias = Embedding(user,1)
        self.movie_factor = Embedding(movie,factors)
        self.movie_bias = Embedding(movie,1)
        self.y_range = y_range
    
    def forward(self,x):
        user = self.user_factor(x[:,0])
        movie = self.movie_factor(x[:,1])
        res = (user * movie).sum(dim=1,keepdim=True)
        res+= self.user_bias(x[:,0]) + self.movie_bias(x[:,1])        
        return sigmoid_range(res,*self.y_range)
```

```python
model = DotProductBias(len(dls.classes['userId']), len(dls.classes['title']),50,(0.5,5.5))
learn = Learner(dls,model,loss_func=MSELossFlat())
learn.fit_one_cycle(5,5e-3)
```

```python
class CollabNN(Module):
    def __init__(self, user_sz, item_sz, y_range=(0,5.5), n_act=100):
        self.user_factors = Embedding(*user_sz)
        self.item_factors = Embedding(*item_sz)
        self.layers = nn.Sequential(
            nn.Linear(user_sz[1]+item_sz[1], n_act),
            nn.ReLU(),
            nn.Linear(n_act, 1))
        self.y_range = y_range
        
    def forward(self, x):
        embs = self.user_factors(x[:,0]),self.item_factors(x[:,1])
        x = self.layers(torch.cat(embs, dim=1))
        return sigmoid_range(x, *self.y_range)
```

```python
model = CollabNN(*embs)
```

```python
learn = Learner(dls, model, loss_func=MSELossFlat())
learn.fit_one_cycle(5, 5e-3, wd=0.1)
```
